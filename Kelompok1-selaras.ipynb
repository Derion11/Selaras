
### 1. Pendaftaran dan Login:
   - **DML**:
     - Buat prosedur untuk memasukkan data pengguna baru ke dalam tabel `user`.
       - Ambil data seperti `username` dan `user_pass`.
       - Pastikan `username` unik.
       - Enkripsi kata sandi sebelum dimasukkan ke dalam tabel.
   - **DDL**:
     - Tentukan struktur tabel `user` dengan kolom-kolom berikut:
       - `u_id`: ID pengguna (auto increment).
       - `username`: Nama pengguna (unik).
       - `user_pass`: Kata sandi pengguna (diacak atau dienkripsi).
       - `user_created`: Waktu pembuatan pengguna (timestamp).
       - `user_deleted`: Tanggal penghapusan pengguna (jika ada).
   - **DQL**:
     - Buat query untuk memeriksa keberadaan pengguna saat login.
     - Verifikasi kata sandi yang dimasukkan dengan kata sandi yang tersimpan setelah mendapatkan pengguna dengan nama pengguna yang cocok.

### 2. Pengelolaan Kegiatan:
   - **DML**:
     - Buat prosedur untuk memasukkan kegiatan baru ke dalam tabel `kegiatan`.
       - Ambil data seperti `id_user`, `jenis`, `nama_kegiatan`, `desc_kegiatan`, `tanggal`, dan `prioritas`.
   - **DDL**:
     - Tentukan struktur tabel `kegiatan` dengan kolom-kolom berikut:
       - `id`: ID kegiatan (auto increment).
       - `id_user`: ID pengguna yang membuat kegiatan (kunci asing ke tabel `user`).
       - `jenis`: Jenis kegiatan (misalnya, m_karir, m_keluarga, m_kamu).
       - `nama_kegiatan`: Nama kegiatan.
       - `desc_kegiatan`: Deskripsi kegiatan.
       - `tanggal`: Tanggal kegiatan.
       - `prioritas`: Prioritas kegiatan (tinggi, sedang, rendah).
   - **DQL**:
     - Buat query untuk mengambil daftar kegiatan pengguna berdasarkan ID pengguna tertentu.
